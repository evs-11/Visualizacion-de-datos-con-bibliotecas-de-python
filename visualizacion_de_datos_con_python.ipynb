# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
# importamos pandas y alamcenamos el df

import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')

# vemos las 5 filas primeras
df.head()
```

```python
# cuantos registro tiene
df.shape
```

```python
# conocer más del df, tipos de datos, nulos etc.
df.info()
```

```python
# configuramos como índice al pais

df.set_index('Pais',inplace=True)
# visualizamos
df
```

```python
# generamos otro df que contenga los años y la cantidad de inmigrantes de Colombia
# generamos una variable que genera una lista con los string de las columnas.
# list para poder mapear, la función de string a un rango que vvaya entre 1980 inclusivo hasta 2014 exclusivo

anos = list(map(str,range(1980,2014)))
# visualizar
anos
```

```python
# aplicar filtro colombia y con el método .loc que va a encontrar colombia dentro del filtro con la variable anos

colombia = df.loc['Colombia',anos]
# muestra los años y a la derecha los inmigrantes
colombia
```

```python
# creamos un df para verlo sencillo y trabajarlo, creamos un df con la esta información, hacemos un diccionario
# diccionario = con el año y el objeto colombia que es pandas series y tomar los índices y pasamos a una lista,
# el 2do. elemento inmigrantes del pandas series colombia los valores y .tolist

col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
# creamos el df datos_col, con el diccionario col_dict

datos_col = pd.DataFrame(col_dict)
# vemos los últimos 5 registros del df recién creado de colombianos hacia Canada
datos_col.tail()
```

###**Para saber más: en qué se relaciona el plano cartesiano con la visualización**

Plano cartesiano. Con el uso del plano cartesiano, constituido por el eje X y el eje Y, es posible crear un sistema de coordenadas bidimensional que facilita la localización y análisis de datos. Estos ejes también son conocidos por los siguientes nombres:

eje x: eje de las abscisas.
eje y: eje de las ordenadas.
Los ejes son componentes esenciales debido a sus combinaciones, que posibilitan una mayor precisión en la identificación de puntos en un espacio bidimensional. En la figura abajo, tenemos la representación de un plano cartesiano conteniendo los ejes X y Y para ejemplificar. Con él podemos notar algunas características importantes:

El eje X es horizontal.
El eje Y es vertical.
En el centro tenemos el origen, el valor 0 (cero).
Hacia la derecha y hacia arriba tenemos valores positivos.
Hacia la izquierda y hacia abajo tenemos valores negativos.
los ejes nos permiten especificar la ubicación exacta de un punto en un espacio bidimensional, de modo que los valores de X y Y son llamados coordenadas porque describen la ubicación del punto en relación al origen. En la figura abajo, tenemos un plano cartesiano conteniendo puntos de algunas coordenadas: (2,5), (1,2), (0,-1), (-1,-4) y (-2,-7). Considera la representación y observa los puntos disponibles. Así puedes visualizar mejor la relación entre ellos y entender patrones y tendencias en los datos.
Cuando creamos un gráfico, las coordenadas de los puntos son utilizadas para definir la posición de cada punto en el gráfico. Y una curiosidad interesante: el plano cartesiano y el eje de coordenadas son ampliamente usados en varios campos, incluyendo Matemáticas, Física, Ingeniería, Ciencia de Datos y muchos otros, siendo una herramienta fundamental para la representación de datos y para la visualización de resultados en diversos contextos.



###**Creando el primer gráfico con matplotlib**

```python
# importar bilblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
# para hacer un gráfico de líneas llamamos a plt. Queremos dibujar años en eje X, luego en el eje Y inmigrantes
# muestra la salida en el espacio de memoria donde quedó almacenada en ese momento, donde se instanció.
# podemos ajustar el tamaño de la figura( ancho por alto)
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
# para mejorar el gráfico con xticks ponemos cada año como string, se hace de forma manual
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
# para evitar que aparezca el espacio de memoria donde quedó almacenada, show muestra la gráfica
plt.show()
```

###**Añadiendo títulos y rótulos**

```python
# todos con el comandos utilizando el módulo plt

plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
# Para colocar el título a lo anterior
plt.title('Inmigración de colombianos hacia Canadá')
# Etiqueta para el eje X
plt.xlabel('Año')
# Etiqueta eje Y
plt.ylabel('Número de Inmigrantes')
# show debe ser el último comando
plt.show()
```

###**Creando una figura**

```python
# Generamos el lienzo(donde voy a graficar) y la gráfica
# se almacena en la variable fig, que es el lienzo y la otra variable ax, que es la gráfica
# el método .subplots, con el tamaño de la figura(el lienzo)
fig, ax = plt.subplots(figsize=(8,4))
# tomamos ax.plot lo que va en el eje X y los datos del eje Y
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
# ax el método .xaxis.set_major_locator, un localizador múltiple(multipleLocator) de 5 en 5 que coloque los ticks de forma automática
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# mostramos
plt.show()
```

###**Añadiendo títulos y rótulos en la figura**

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# colocamos el título en dos líneas con \n
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
# colocamos el eje x
ax.set_xlabel('Año')
# colocamos el eje y
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

###**Creando subplots en una dirección**

```python
# gráfico Boxsplot, para estadistica descriptiva . Crear lienzo almacenada en variable fig axs porque son 2 figuras.
# almacenamos en subplots, con 1 fila con 2 columnas para subgráficos y tamaño del lienzo
fig, axs = plt.subplots(1,2,figsize=(10,4))
# la primera gráfica 0.plot los datos
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
# el título
axs[0].set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
# el rótulo eje x
axs[0].set_xlabel('Año')
# el rótulo eje y
axs[0].set_ylabel('Número de Inmigrantes')
# especificar el rango de los ticks
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
# ponemos una cuadricula,
axs[0].grid()

# del otro gráfico, es boxplot(diagrama de caja) mide una variable, aquí queremos medir la cantidad de colombianos
axs[1].boxplot(datos_col['Inmigrantes'])
# el título
axs[1].set_title('Boxplot de la inmigración de colombianos\nhacia Canadá 1980 - 2013')
# el rótulo eje x
axs[1].set_xlabel('Colombia')
# el rótulo eje y
axs[1].set_ylabel('Número de Inmigrantes')
# ponemos una cuadricula,
axs[1].grid()

# en el 2do. muestra el 1er. cuartil que sería alrededor de 200 inmigrantes, en el 2do. cuartil o sea el 50 % de la muestra, alrededor de 600, es la mediana
# raya anaranjada, luego el 75 % alrededor de 4100
plt.show()
```

###**Creando subplots en dos direcciones #1**

```python
# No sólo en sentido horizontal. Analizamos otros paises
# 2,2 son dos filas y dos columnas o sea 4 gráficos
# se cambia el espacio entre gráficos con .subplots_adjust es h(altura) y w(ancho) space
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
# la primera gráfica 0,0.plot los datos de la 1era. fila y 1ra. columna. Del df original .loc(Colombia)los años.
axs[0,0].plot(df.loc['Colombia',anos])
# el título
axs[0,0].set_title('Colombia')


# la primera gráfica 0,1.plot los datos de la 1era. fila y 1ra. columna. Del df original .loc(Colombia)los años.
axs[0,1].plot(df.loc['Brasil',anos])
# el título
axs[0,1].set_title('Brasil')

# la primera gráfica 1,0.plot los datos de la 1era. fila y 1ra. columna. Del df original .loc(Colombia)los años.
axs[1,0].plot(df.loc['Argentina',anos])
# el título
axs[1,0].set_title('Argentina')

# la primera gráfica 1,1.plot los datos de la 1era. fila y 1ra. columna. Del df original .loc(Colombia)los años.
axs[1,1].plot(df.loc['Perú',anos])
# el título
axs[1,1].set_title('Perú')

# los gráficos tienen los mismos ejes, hay que hacerles el mismo ajuste en los años. Con un ciclo de lazos for
# con .flat permite iterar en las gráficas y hacer las siguientes operaciones

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')

plt.show()
```
